# Analysis of Diifference between poorest and richest province in China by time

```python
def find_highest_lowest(Years):
    lower_GDP = df_trans.loc[:,[Years]].min().to_numpy()[0]
    highest_GDP = df_trans.loc[:,[Years]].max().to_numpy()[0]   
    print(f"The lowest GDP in {Years}: {df_trans[df_trans[Years] == lower_GDP].index.values}") 
    print(f"The highest GDP in {Years}: {df_trans[df_trans[Years] == highest_GDP].index.values}")
    return lower_GDP, highest_GDP


def find_the_difference(Years):
    lower_GDP, highest_GDP = find_highest_lowest(Years)
    
    difference_GDP = highest_GDP - lower_GDP
    
    return difference_GDP
```

```python
diff_1992_2022 = [find_the_difference(1992), find_the_difference(2020)]

plt.title("GDP Comparaison between the poorest and the richest province for 1992 and 2022 ")
plt.ylabel("Gross domestic product")
sns.barplot(x= ["1992", "2020"] , y = diff_1992_2022 )
```
